The above figure represents a two layered neural network(Note:neural networks can have other architectures than the on showed above). We are going to use the above architecture at times called a fully connected layer or a dense layer for this exercise, few properties of the network are as follows:

###NEURON

Each circle in the above figure are usually called a neuron and it takes values from all the previous neurons and pass it through an activation function.

The neurons with the same colour belong to the same layer. The more the layers the better the model can learn complex features.

The input layer is the very first layer of the neural network and has the same amount of neurons as the features(data dimesion).

The hidden layer is the layers between the input layer and output layer.

The output layer will have the same number of neurons as the expected output dimension.


###SYNAPSE

This is the connection between neurons and each connection is weghted output of the orgin neuron.

#KERAS

As you might have seen, it is not easy to perform the above calculations each time you need to make a neural network, luckily for you there are many high level libraries out there to help you with making neural networks, lets have a look at one such library which is called keras. Recently the library called tensorflow by google has taken over keras.


````
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
mnist = tf.keras.datasets.mnist
````

Since the keras library already has the links and commands to download our dataset, we just have to call the function for it.

````
(x_train, y_train),(x_test, y_test) = mnist.load_data()
````

Our dataset looks something like this.

````
'''
This Code is for visualization, understanding this is not necessary
'''
fig = plt.figure()
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.tight_layout()
  plt.imshow(x_train[i], cmap='gray', interpolation='none')
  plt.title("Digit: {}".format(y_train[i]))
  plt.xticks([])
  plt.yticks([])
fig
````

The Sequential class in keras is what we can call a blank slate, here we can stack neural network layers. You can learn more about this [here](https://www.tensorflow.org/api_docs/python/tf/keras/models/Sequential).

````
model = tf.keras.models.Sequential()
````

the variable mmodel is an object of sequential type and it has many in-biult function which we can use to make our lives easy. 


```
model.add(())
```

This will add a neural network layer to the stack.Our dataset is an 28 x 28 image we first flatten it meaning we convert into a single 784 unit input.


````
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
````

The Dense method can be used to add hidden units, it takes argument as the number of neuron required in that layer and also accepts other arguments which you can refer in the documentation. We will also need an activation to make our network learn non linear relationships.

````
model.add(tf.keras.layers.Dense(128 # 128 neurons
                                , activation = 'relu'))
model.add(tf.keras.layers.Dense(128, activation = 'relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
````


To assign optimizer,loss function we use
```
model.compile()
```

We are using adam as the optimizer you can read more about it on the links provided in Week 3, sparse_categorical_crossentropy is the entropy loss you may refer it [here](https://towardsdatascience.com/understanding-binary-cross-entropy-log-loss-a-visual-explanation-a3ac6025181a).


````
model.compile(optimizer='adam',# Refer online on the details of adam and other omptimizers
              loss='sparse_categorical_crossentropy', # Entropy loss you can refer it here https://gombru.github.io/2018/05/23/cross_entropy_loss/
              metrics=['accuracy'])
````

Normalizing the input is a good practise doing so will let the network understand that our values can only be between 0 and 1. This speeds up the training, try with and without this.

````
#x_train = x_train / 255  
````

You have everything you need now, calling 

```
model.fit
```
with the x_train and y_train as arguments it will do the magic, the forward pass and the backpropagation. If this is slow you can go to *Runtime* in menu bar and choose *Change Runtime*, in that change the option *None* to *GPU*.

````
model.fit(x_train, y_train, epochs=10)
````

#EVALUATE YOUR MODEL

Wait! its not over we have to look at how good our model is performing.

````
score = model.evaluate(x_test, y_test, verbose=0)
print('Test score:',score[0])
print('Test accuracy:',score[1])
````

Looks like our model is performing really well, 97% accurte on the hand written characters. But I would like to know which data points its doing well and which its doing badly on.

````
# The predict_classes function outputs the highest probability class
# according to the trained classifier for each input example.
predicted_classes = model.predict_classes(x_test)

# Check which items we got right / wrong
correct_indices = np.nonzero(predicted_classes == y_test)[0]
incorrect_indices = np.nonzero(predicted_classes != y_test)[0]
````

The data samples our model got correct on.

````
'''
This Code is for visualization, understanding this is not necessary
'''

fig = plt.figure(figsize=(10, 10))
for i,correct in enumerate(correct_indices[:25]):
  plt.subplot(5,5,i+1)
  plt.tight_layout()
  plt.imshow(x_test[correct], cmap='gray', interpolation='none')
  plt.title("Predicted: {}".format(y_test[correct]))
  plt.xticks([])
  plt.yticks([])
````

The data samples the model failed to predict on,

````
'''
This Code is for visualization, understanding this is not necessary
'''

fig = plt.figure(figsize=(10, 10))
for i,incorrect in enumerate(incorrect_indices[:25]):
  plt.subplot(5,5,i+1)
  plt.tight_layout()
  plt.imshow(x_test[incorrect], cmap='gray', interpolation='none')
  plt.title("Predicted {}, True {}".format(predicted_classes[incorrect], y_test[incorrect]))
  plt.xticks([])
  plt.yticks([])
fig
````
