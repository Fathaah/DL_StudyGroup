#numpy

Below are some more arithmetic stuffs I have done with numpy.

```
# This creates an numpy array of 20 elements from normal distribution with
# mean = 10    standard deviation = 3
# FOR FIRST YEAR STUDENTS: YOU WILL STUDY ABOUT DISTRIBUTIONS IN S3
# SO DON'T WORRY SEEING THOSE FANCY TERMS LIKE GAUSSIAN DISTRIBUTION, etc.

a = np.random.normal(10,3,20)
a
```

```
# This creates an numpy array of 20 elements from uniform distribution with
# a = 5 , b = 10

b = np.random.uniform(5,10,20)
b
```

```
# ADDING OR SUBTRACTING BOTH THESE ARRAYS ARE SIMPLE

a+b
```

```
# IF YOU WANT TO ADD 5 TO ALL ELEMENTS IN THE FIRST ARRAY

a+5
```

```
# LET'S LOOK AT TAKING DOT PRODUCTS OF THESE TWO ARRAYS

np.dot(a,b)
```

```
# NOW TAKING sine OF THE SECOND ARRAY AND STORING IT IN A VARIABLE z

z = np.sin(b)
z
```

```
# PASSING OUR ARRAY z TO A FUNCTION f(z) = 1/1+e^(-z)

f = 1/(1+np.exp(-z))
f

# THAT WAS SIMPLE RIGHT?
# THE FUNCTION f(z) = 1/1+e^(-z) IS CALLED SIGMOID FUNCTION.
# IT IS ONE OF THE ACTIVATION FUNCTIONS USED IN DEEP LEARNING.
```

```
# IF YOU WANT TO CONVERT THE ABOVE ARRAY INTO A 4x5 MATRIX

f = np.reshape(f,(4,5))
f
```

```
# LET'S CONVERT ARRAY a INTO A MATRIX OF 5x4 AND DO MATRIX MULTIPLICATION WITH f

a = np.reshape(a,(5,4))

np.matmul(a,f)
```

That's a 5x5 matrix you got after matrix multiplication.

There are many other useful functions. You will come across other useful functions in the future.

```
irisdf = pd.read_csv("iris.data",names=["sepal length",
                                        "sepal width",
                                        "petal length",
                                        "petal width",
                                        "Iris class"])
irisdf
```

```
irisdf["Iris class"].unique()
```

Here "Iris class" is categorical with string values which may give you a headache while plotting. And you have to always convert a string categorical column to integers or you have to use One Hot Encode before feeding it into a ML or DL model.<br><br>
I think I should take another burden.

```
def to_categorical(value):
  if value=="Iris-setosa":
    return 1
  elif value=="Iris-versicolor":
    return 2
  else:
    return 3

# THE ABOVE FUNCTION JUST ASSIGNS 1,2 OR 3 ACCORDING TO THE IRIS CLASS
# THE CODE IS QUITE OBVIOUS I THINK

irisdf["Iris class"] = irisdf["Iris class"].apply(to_categorical)
irisdf
```
