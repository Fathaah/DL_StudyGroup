## Data types in Python

You may have noticed that some numbers we've stored inside varaibles contained decimal spaces while others do not and some where words and even sentences

In python there are different **Types** of data that can be stored in variables

**Python Number**

There are three types of python numbers.

  * **Integers** Integers in python are any numbers without a decimal point. They can be of any length. Integers fall under int python class. For example, 1, 100, 9999, etc.
    
  * **Floating Point** The floating point numbers contain decimal points and integer part and floating points are separated by the dot (.). Python floating point numbers support up to 15 decimal places. For example, 1.192, 20.0, 99.9999, etc.
  * **Complex** Complex numbers in python are in the form of x + yj. Here, x is the real part and y is the imaginary part. For example: 4 + 5j, 1 + 2j, etc.
  
  
**Python String**

The python strings are the sequences of characters within single or double quotes (`or`). Triple quotes (`, `) can be used for multi-line strings. A single character in the python is also represented as a string, unlike C or C++ which have separate data type to store single characters.

**Heres a couple of examples, you can play around with these **

````python
a = 100 # assigning integer 100 to a variable a
b = 99.99 # assigning floating point number 99.99 to a variable b
c = 3 + 2j # assigning complex number 3 + 2j to a variable c

print(a+b, c+a, b+c)


a = 100 # assigning integer 100 to a variable a
b = 99.99 # assigning floating point number 99.99 to a variable b
c = 3 + 2j # assigning complex number 3 + 2j to a variable c


print(a, b, c)
````

Conditional statements lets your program execute different code based on a condition.

Consider(RUN) the following code:

````python
for i in [ 1, 2, 3, 4, 5] :
    print(i)
````

The above code utilises a **for loop** to **iterate** through every value in the given list. The variable i( you can use any name) takes on each value in the list consecutively, executing the same block of code for every value

## While loop

With the while loop we can execute a set of statements as long as a condition is true.

````python
i = 1
while i < 6:
  print(i)
  i += 1
````

** Watch out for infinite loops!**

In the above code, notice that the variable( i ) used in the condition ( i < 6 ) is changed with each execution, until eventually the condition returns false, exiting from the loop. what happens when this is not the case?

````python
i = 1
while i < 6:
    print(i)
    i = i + 1 #REMOVE THIS LINE OF CODE
````

As you may have observed, with the last line removed, the condition remains True always, hence creating an **Infinite Loop** . You can forcefully terminate the program with stop button in the tool bar
